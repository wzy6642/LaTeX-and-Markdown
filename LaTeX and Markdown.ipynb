# LaTeX公式排版教程

## 公式编号
* 人工编号的方式可以在公式后边添加`\label{eq:1}`进行编号，这里的`1`就是给**公式编号**，代码如下：
```latex
\begin{equation}\label{eq:1}
x^n+y^n=z^n
\end{equation}```
\begin{equation}\label{eq:1}
x^n+y^n=z^n
\end{equation}
* 在Latex公式，使用`\ `和`\; `来表示**空格**，代码如下：
```latex
\begin{equation}\label{eq:2}
\epsilon>0 \;  a=b
\end{equation}```
\begin{equation}\label{eq:2}
\epsilon>0 \;  a=b
\end{equation}
* 在Latex公式，使用`\\`来表示**换行**，代码如下：
```latex
\begin{equation}\label{eq:3}
\sum_{i=1}^n p_i\\
\textrm{另起一行}
\end{equation}```
\begin{equation}\label{eq:3}
\sum_{i=1}^n p_i\\
\textrm{另起一行}
\end{equation}
* 在Latex公式，使用`\ref`来表示**公式引用**，代码如下：
```latex
\begin{equation}\label{eq:4}
\frac{a}{b}\\
\end{equation}
从 (\ref{eq:4}), 我们可以看出是一个分式```
\begin{equation}\label{eq:4}
\frac{a}{b}\\
\end{equation}
从 (\ref{eq:4}), 我们可以看出是一个分式

## 希腊字母
* 在Latex公式，常用的**小写希腊字母**，代码如下：
```latex
\begin{equation}\label{eq:5}
\alpha,\beta,\gamma,\delta,\epsilon,\varepsilon,\zeta,\eta,\theta,\vartheta,\iota,\kappa,\\
\lambda,\mu,\nu,\xi,o,\pi,\varpi,\rho,\varrho,\sigma,\varsigma,\tau,\upsilon,\phi,\varphi,\chi,\psi,\omega
\end{equation}```
\begin{equation}\label{eq:5}
\alpha,\beta,\gamma,\delta,\epsilon,\varepsilon,\zeta,\eta,\theta,\vartheta,\iota,\kappa,\\
\lambda,\mu,\nu,\xi,o,\pi,\varpi,\rho,\varrho,\sigma,\varsigma,\tau,\upsilon,\phi,\varphi,\chi,\psi,\omega
\end{equation}
* 在Latex公式，常用的**大写希腊字母**，代码如下：
```latex
\begin{equation}\label{eq:6}
\Gamma,\Delta,\Theta,\Lambda,\Xi,\Pi,\Sigma,\Upsilon,\Phi,\Psi,\Omega
\end{equation}```
\begin{equation}\label{eq:6}
\Gamma,\Delta,\Theta,\Lambda,\Xi,\Pi,\Sigma,\Upsilon,\Phi,\Psi,\Omega
\end{equation}
* 在Latex公式，常用的**重音符**，代码如下：
```latex
\begin{equation}\label{eq:7}
\hat{a},\check{a},\tilde{a},\acute{a},\bar{x},\grave{x},\breve{x},\ddot{x},\dot{x}
\end{equation}```
\begin{equation}\label{eq:7}
\hat{a},\check{a},\tilde{a},\acute{a},\bar{x},\grave{x},\breve{x},\ddot{x},\dot{x}
\end{equation}

## 上标下标
* 在Latex公式，**上标**使用`^`，**下标**使用`_`，代码如下：
```latex
\begin{equation}\label{eq:8}
a^2+L_a=doubi
\end{equation}```
\begin{equation}\label{eq:8}
a^2+L_a=doubi
\end{equation}
* 如果上标或下标后需要添加**多个字符**，那么请使用`花括号`把多个字符放一起，代码如下：
```latex
\begin{equation}\label{eq:9}
a^{x+y}+L_{dexi} = doubi
\end{equation}```
\begin{equation}\label{eq:9}
a^{x+y}+L_{dexi} = doubi
\end{equation}

## 平方根
* 输入命令为：`\sqrt`，n 次方根相应地为: `\sqrt[n]`，如果只需要符号请使用`\surd`，代码如下：
```latex
\begin{equation}\label{eq:10}
\sqrt{a^2}-\sqrt[5]{a}=\surd{a}
\end{equation}```
\begin{equation}\label{eq:10}
\sqrt{a^2}-\sqrt[5]{a}=\surd{a}
\end{equation}

## 大括号
* 使用`\overbrace`和`\underbrace`可以在**表达式上方、下方添加花括号**，代码如下：
```latex
\begin{equation}\label{eq:11}
\overbrace{1,2,3,4,5,6,7}_{前几个数字} + \underbrace{8,9,10,11,12}_{后几个数字}
\end{equation}```
\begin{equation}\label{eq:11}
\overbrace{1,2,3,4,5,6,7}_{前几个数字} + \underbrace{8,9,10,11,12}_{后几个数字}
\end{equation}

## 向量
* **向量**可以使用`\vec`，如果希望定义从左向右，请使用`\overleftarrow`，代码如下：
```latex
\begin{equation}\label{eq:12}
\vec a,\ \vec{ab},\ \overleftarrow{ab},\ \overrightarrow{lindexi}
\end{equation}```
\begin{equation}\label{eq:12}
\vec a,\ \vec{ab},\ \overleftarrow{ab},\ \overrightarrow{lindexi}
\end{equation}

## 圆点
* 在数学，经常使用圆点做**点乘**，点在 Latex 使用 `\cdot` 如表示 x·y 可以使用下面代码：
```latex
\begin{equation}\label{eq:13}
x\cdot y
\end{equation}```
\begin{equation}\label{eq:13}
x\cdot y
\end{equation}

## 三角函数
* 常用的**三角函数**表示方法如下列代码所示：
```latex
\begin{equation}\label{eq:14}
\sin\theta,\ \cos\theta,\ \tan\theta,\ \arcsin\frac{L}{i},\ \arccos\frac{n}{D},\ \arctan\frac{e}{x},\ \sinh l,\ \cosh i,\\
\tanh n,\ \operatorname{sh}d,\ \operatorname{argsh}e,\ \operatorname{ch}x,\ \operatorname{argch}i,\ \operatorname{th}i\\
\operatorname{argth}m,\ \lim_{\Delta x\to 0}x,\ \max H,\ \min L,\ \inf s,\ \sup t,\ \exp\!t,\ \ln X,\ \lg X,\ \log X\\
\log_\alpha X,\ \ker x,\ \deg x,\ \gcd(T,U,V,W,X),\ \Pr x,\ \det x,\ \hom x,\ \arg x,\ \dim x,\ \lim_{t\to n}T
\end{equation}```
\begin{equation}\label{eq:14}
\sin\theta,\ \cos\theta,\ \tan\theta,\ \arcsin\frac{L}{i},\ \arccos\frac{n}{D},\ \arctan\frac{e}{x},\ \sinh l,\ \cosh i,\\ 
\tanh n,\ \operatorname{sh}d,\ \operatorname{argsh}e,\ \operatorname{ch}x,\ \operatorname{argch}i,\ \operatorname{th}i\\
\operatorname{argth}m,\ \lim_{\Delta x\to 0}x,\ \max H,\ \min L,\ \inf s,\ \sup t,\ \exp\!t,\ \ln X,\ \lg X,\ \log X\\
\log_\alpha X,\ \ker x,\ \deg x,\ \gcd(T,U,V,W,X),\ \Pr x,\ \det x,\ \hom x,\ \arg x,\ \dim x,\ \lim_{t\to n}T
\end{equation}

## 极限
* 常用的**极限**表示方法如下代码所示：
```latex
\begin{equation}\label{eq:15}
\lim_{x\to 0} \frac{\sin x}{x}=1
\end{equation}```
\begin{equation}\label{eq:15}
\lim_{x\to 0} \frac{\sin x}{x}=1
\end{equation}

## 微分
* 常用的**微分**表示方法如下代码所示：
```latex
\begin{equation}\label{eq:16}
\pmod{m},a \bmod b,\nabla,\partial x,\mathrm{d}x,\dot x,\ddot y
\end{equation}```
\begin{equation}\label{eq:16}
\pmod{m},a \bmod b,\nabla,\partial x,\mathrm{d}x,\dot x,\ddot y
\end{equation}

## 分数
* **分数**使用 `\frac` 后面使用两个花括号，第一个是上标，第二个是下标，代码如下：
```latex
\begin{equation}\label{eq:17}
y=\frac{x}{b}
\end{equation}```
\begin{equation}\label{eq:17}
y=\frac{x}{b}
\end{equation}

## 二项系数
* **二线系数**可以使用 `{ 第一项 \choose 第二项 }` 或 `{ 第一项 \atop 第二项 }`，*注意，第一项 和括号需要有空格，第二项和括号要有空格*，代码如下：
```latex
\begin{equation}\label{eq:18}
{ 第一项 \choose 第二项 }\\ { 第一项 \atop 第二项 }
\end{equation}```
\begin{equation}\label{eq:18}
{ 第一项 \choose 第二项 }\\ { 第一项 \atop 第二项 }
\end{equation}

## 运算

### 积分
+ **积分**符号`\int`，代码如下：
```latex
\begin{equation}\label{eq:19}
\int_{0}^{x+2}{k*3}
\end{equation}```
\begin{equation}\label{eq:19}
\int_{0}^{x+2}{k*3}
\end{equation}

### 求和
+ **求和**运算符 `\sum`，代码如下：
```latex
\begin{equation}\label{eq:20}
\sum_{i=1}^{n}{i}
\end{equation}```
\begin{equation}\label{eq:20}
\sum_{i=1}^{n}{i}
\end{equation}

### 乘积
+ **乘积**运算符 `\prod`，代码如下：
```latex
\begin{equation}\label{eq:21}
\prod_{2}^{n}{\epsilon}
\end{equation}```
\begin{equation}\label{eq:21}
\prod_{2}^{n}{\epsilon}
\end{equation}

## 括号
* 圆括号、方括号都可以直接输入，花括号`\{`，括号的大小可以使用`\big`, `\Big`, `\bigg` 及 `\Bigg` 表示不同的大小，代码如下：
```latex
\begin{equation}\label{eq:22}
(\ ),[\ ],\{\ \},\big(\ \big),\Big(\ \Big),\bigg(\ \bigg),\Bigg(\ \Bigg)
\end{equation}```
\begin{equation}\label{eq:22}
(\ ),[\ ],\{\ \},\big(\ \big),\Big(\ \Big),\bigg(\ \bigg),\Bigg(\ \Bigg)
\end{equation}

## 点
* 在输入省略号就经常需要使用**点**，`\ldots`使用的是对其下方，`\cdots`是居中，`\vdots`垂直，`\ddots`对角，代码如下：
```latex
\begin{equation}\label{eq:23}
lindexi \ldots doubi\\lindexi \vdots doubi\\lindexi \ddots doubi\\lindexi \cdots doubi
\end{equation}```
\begin{equation}\label{eq:23}
lindexi \ldots doubi\\lindexi \vdots doubi\\lindexi \ddots doubi\\lindexi \cdots doubi
\end{equation}

## 数组
* 使用`array`表示**数组**，代码如下：
```latex
\begin{equation}\label{eq:24}
\mathbf{X} = 
    \left( 
        \begin{array}
            {ccc} x_{11} & x_{12} & \ldots \\ x_{21} & x_{22} & \ldots \\ \vdots & \vdots & \ddots 
        \end{array} 
    \right)
\end{equation}```
\begin{equation}\label{eq:24}
\mathbf{X} = 
    \left( 
        \begin{array}
            {ccc} x_{11} & x_{12} & \ldots \\ x_{21} & x_{22} & \ldots \\ \vdots & \vdots & \ddots 
        \end{array} 
    \right)
\end{equation}

## 二元关系符

### 不等号
* 使用`>`表示**大于**，使用`<`表示**小于**，使用`=`表示**等于**，使用`\leq`或者`\le`表示**小于等于**，使用`\geq`或者`\ge`表示**大于等于**，使用`\equiv`表示**恒等于**，使用`\sim`表示**相似**，使用`\approx`表示**约等于**，使用`\neq`表示**不等于**，代码如下：
```latex
\begin{equation}\label{eq:25}
a \leq b \le a \\ a \geq b \ge a \\ a \equiv b \\ a \sim b \\ a \approx b \\ a \neq b
\end{equation}```
\begin{equation}\label{eq:25}
a \leq b \le a \\ a \geq b \ge a \\ a \equiv b \\ a \sim b \\ a \approx b \\ a \neq b
\end{equation}

### 集合
* 使用`\subset`表示**子集**，使用`\subseteq`表示**真子集**，使用`\supset`表示**包括**，使用`\supseteq`表示**真包括**，使用`\emptyset`表示**空集**，使用`\cap`，`\bigcap`表示**交**，使用`\cup`，`\bigcup`表示**并**，代码如下：
```latex
\begin{equation}\label{eq:26}
A \subset B \\ A \subseteq B \\ A \supset B \\ A \supseteq B \\ \emptyset \\ A \cap B , A \bigcap B \\ A \cup B , A \bigcup B
\end{equation}```
\begin{equation}\label{eq:26}
A \subset B \\ A \subseteq B \\ A \supset B \\ A \supseteq B \\ \emptyset \\ A \cap B , A \bigcap B \\ A \cup B , A \bigcup B
\end{equation}

### 元素与集合
* 使用`\in`，`\ni`表示**包含**，使用`\notin`表示**不包含**，代码如下：
```latex
\begin{equation}\label{eq:27}
a \in A \\ A \ni a \\ a \notin A
\end{equation}```
\begin{equation}\label{eq:27}
a \in A \\ A \ni a \\ a \notin A
\end{equation}

### 命题
* 使用`\forall`表示**全称命题**，使用`\exists`表示**特称命题**，代码如下：
```latex
\begin{equation}\label{eq:28}
\forall A \\ \exists A
\end{equation}```
\begin{equation}\label{eq:28}
\forall A \\ \exists A
\end{equation}

### 逻辑
* 使用`\land`，`\wedge`，`\bigwedge`表示**且**，使用`\lor`，`\vee`，`\bigvee`表示**或**，使用`\lnot`，`\neg`表示**非**，使用`\setminus`，`\backslash`表示**两个集合的差运算**，代码如下：
```latex
\begin{equation}\label{eq:29}
\land,\wedge,\bigwedge \\ \bar{q} \to p \\ \lor,\vee,\bigvee \\ \lnot,\neg q \\ A \setminus B,A \backslash B
\end{equation}```
\begin{equation}\label{eq:29}
\land,\wedge,\bigwedge \\ \bar{q} \to p \\ \lor,\vee,\bigvee \\ \lnot,\neg q \\ A \setminus B,A \backslash B
\end{equation}

## 常用二元运算符
* 常用二元运算符代码如下：
```latex
\begin{equation}\label{eq:30}
\textrm{加}+ \ ，\textrm{加减}\pm \ ，\textrm{点乘}\cdot \ ，\textrm{叉乘}\times \ ，\textrm{减}- \ ，\textrm{除号}\div \\
\textrm{斜杠除}\setminus 
\end{equation}```
\begin{equation}\label{eq:30}
\textrm{加}+ \ ，\textrm{加减}\pm \ ，\textrm{点乘}\cdot \ ，\textrm{叉乘}\times \ ，\textrm{减}- \ ，\textrm{除号}\div \\
\textrm{斜杠除}\setminus 
\end{equation}

## 常用大尺寸运算符
* 常用大尺寸运算符代码如下：
```latex
\begin{equation}\label{eq:31}
\textrm{求和}\sum \ ，\textrm{环积分}\oint \ ，\textrm{积分}\int \ ，\textrm{连乘}\prod 
\end{equation}```
\begin{equation}\label{eq:31}
\textrm{求和}\sum \ ，\textrm{环积分}\oint \ ，\textrm{积分}\int \ ，\textrm{连乘}\prod 
\end{equation}

## 常用箭头
* 常用箭头代码如下：
```latex
\begin{equation}\label{eq:32}
\textrm{必要不充分}\Leftarrow \ ，\textrm{充分不必要}\Rightarrow \ ，\textrm{充要}\Leftrightarrow 
\end{equation}```
\begin{equation}\label{eq:32}
\textrm{必要不充分}\Leftarrow \ ，\textrm{充分不必要}\Rightarrow \ ，\textrm{充要}\Leftrightarrow 
\end{equation}
***

# Markdown 教程

## Markdown 标题
* 使用 `#` 号可表示 1-6 级**标题**，一级标题对应一个 # 号，二级标题对应两个 # 号，以此类推。
\begin{example}\label{example:1}
# 一级标题  ## 二级标题  ### 三级标题  #### 四级标题  ##### 五级标题  ###### 六级标题
\end{example}

## Markdown 段落 
* **段落的换行**是使用两个以上空格加上回车。  
The stock plate division is to use the technology of data mining to extract its various features and then cluster stocks.   
Most of the stock clustering algorithm only extract common indicators of stock investment which cannot describe the stock data well.  

### Markdown 字体
* Markdown 可以使用以下几种**字体**：
```Markdown
*斜体文本*  _斜体文本_  **粗体文本**  __粗体文本__  ***粗斜体文本***  ___粗斜体文本___
```

### Markdown 分隔线

### Markdown 删除线
